```python
import numpy as np
import matplotlib.pyplot as plt
from random import random, seed

seed(0)  

Np = 10000
Nf = 100
Px = np.zeros(Np)
P = np.zeros((Np, Nf))
D = 0.5  

for time in range(Nf):
    for p in range(Np):
        R = random()
        if R <= 0.5:
            Px[p] += 1
        else:
            Px[p] -= 1
            
    P[:, time] = Px
    
    avePx = np.mean(Px)
    varPx = np.mean(Px*Px) - avePx**2

    
    if time % 20 == 0:
        plt.figure(figsize=(10, 6))
        
        
        plt.hist(Px, bins=50, density=True, alpha=0.5, label="Particle distribution")
        
        x_values = np.linspace(min(Px), max(Px), 400)
        gaussian = (1/np.sqrt(4*np.pi*D*time)) * np.exp(-x_values**2/(4*D*time))
        plt.plot(x_values, gaussian, 'r-', label="Gaussian distribution")
        
        plt.title(f"Time frame: {time}")
        plt.xlabel("Position")
        plt.ylabel("Density")
        plt.legend()
        plt.show()
```
